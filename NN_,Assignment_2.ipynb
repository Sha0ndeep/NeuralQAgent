```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from keras.optimizers import Adam
from tensorflow.keras.layers import BatchNormalization
```

```python
train = pd.read_csv('/content/Train82.csv')
```

```python
train_cleaned = train.dropna()
```

```python
X = train_cleaned.drop(columns=['label', 'ImgPath'])
y = train_cleaned['label']
```

```python
X.head()
```

```python
y.head()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=50)
```

```python
#The model
model = Sequential()
model.add(Dense(256, input_dim=X_train.shape[1], activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(len(y.unique()), activation='softmax'))
```

```python
#Compiling
model.compile(Adam(learning_rate=0.0001),loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
trained = model.fit(X_train, y_train, epochs=100, batch_size=64, validation_data=(X_val, y_val))
```

```python
test_df = pd.read_csv('/content/Test82 2.csv')
```

```python
test_cleaned = test_df.dropna()
```

```python
X_test = test_cleaned.drop(columns=['ImgPath'])
X_test_scaled = scaler.transform(X_test)
```

```python
predictions = model.predict(X_test_scaled)
predicted_labels = np.argmax(predictions, axis=1)
```

```python
results_df = test_cleaned[['ImgPath']].copy()
results_df['label'] = predicted_labels
```

```python
results_df.to_csv('/content/output_predictions.csv', index=False)
```
